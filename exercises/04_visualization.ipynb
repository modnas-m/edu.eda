```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

sns.set_theme(style="whitegrid")

# TODO: Replace with your city slug and display name
CITY_SLUG = "종로구"       # <-- change this!
CITY_NAME = "종로구"   # <-- for chart titles

df = pd.read_csv(f"../data/clean_{CITY_SLUG}.csv")
print(f"Loaded {len(df)} rows")
```

---

### 🎨 Part A: matplotlib + seaborn (static charts)

#### Chart 1: Top 15 amenity types (pre-built — study how it works)

```python
# Pre-built: horizontal bar chart of top 15 amenity types
top_amenities = df["amenity"].value_counts().head(15)

fig, ax = plt.subplots(figsize=(10, 6))
top_amenities.plot(kind="barh", ax=ax, color="steelblue")
ax.set_xlabel("Count")
ax.set_ylabel("Amenity Type")
ax.set_title(f"Top 15 Amenity Types in {CITY_NAME}")
ax.invert_yaxis()  # Largest at top
plt.tight_layout()
plt.show()
```

#### Chart 2: Data completeness — **TODO** build this chart

```python
# Create a horizontal bar chart showing % non-null for key columns
# Key columns to check: amenity, primary_name, cuisine, opening_hours, phone, website

key_cols = ["amenity", "primary_name", "cuisine", "opening_hours", "phone", "website"]
available_cols = [c for c in key_cols if c in df.columns]
completeness = df[available_cols].notna().mean() * 100

fig, ax = plt.subplots(figsize=(8, 5))
completeness.sort_values().plot(kind="barh", ax=ax, color="coral")
ax.set_xlabel("% Non-null")
ax.set_title(f"Data Completeness in {CITY_NAME}")
ax.axvline(x=50, color="red", linestyle="--", alpha=0.5, label="50%")
ax.legend()
plt.tight_layout()
plt.show()
```

#### Chart 3: Cuisine distribution — **TODO** bar chart of top 10 `cuisine_clean`

```python
# Bar chart of top 10 cuisines
# Check if cuisine_clean exists, otherwise use cuisine
cuisine_col = "cuisine_clean" if "cuisine_clean" in df.columns else "cuisine"

top_cuisines = df[cuisine_col].value_counts().head(10)

fig, ax = plt.subplots(figsize=(10, 6))
top_cuisines.plot(kind="barh", ax=ax, color="mediumseagreen")
ax.set_xlabel("Count")
ax.set_ylabel("Cuisine Type")
ax.set_title(f"Top 10 Cuisines in {CITY_NAME}")
ax.invert_yaxis()  # Largest at top
plt.tight_layout()
plt.show()
```

### 🗺️ Part B: plotly (interactive maps)

You get the goal — you write the code. See [plotly express docs](https://plotly.com/python/plotly-express/).

#### Map 1: All amenities colored by `amenity_group`, hover shows name

Use `px.scatter_mapbox()` · `mapbox_style="carto-positron"` · `zoom=13`

```python
# Create interactive map of all amenities
# Calculate center of map
center_lat = df["lat"].mean()
center_lon = df["lon"].mean()

fig = px.scatter_mapbox(
    df,
    lat="lat",
    lon="lon",
    color="amenity",
    hover_data={"lat": ":.4f", "lon": ":.4f", "amenity": True, "cuisine": True},
    title=f"All Amenities in {CITY_NAME}",
    mapbox_style="carto-positron",
    zoom=13,
    center={"lat": center_lat, "lon": center_lon},
    height=600
)

fig
```

#### Map 2: Restaurants/cafés/fast_food colored by `cuisine_clean`

```python
# Create restaurant map colored by cuisine
# Filter to food-related amenities
food_amenities = ["restaurant", "cafe", "fast_food", "bar", "pub", "food_court", "bakery"]
df_food = df[df["amenity"].isin(food_amenities)].copy()

print(f"Found {len(df_food)} food establishments")

# Calculate center
center_lat = df_food["lat"].mean()
center_lon = df_food["lon"].mean()

fig = px.scatter_mapbox(
    df_food,
    lat="lat",
    lon="lon",
    color="cuisine",
    hover_data={"lat": ":.4f", "lon": ":.4f", "amenity": True, "cuisine": True},
    title=f"Restaurants & Food Venues by Cuisine in {CITY_NAME}",
    mapbox_style="carto-positron",
    zoom=13,
    center={"lat": center_lat, "lon": center_lon},
    height=600
)

fig
```

#### 💡 Chart 4: Your own question

*My question: Where do different amenity types cluster geographically? Do restaurants/cafés concentrate in specific areas?*

```python
# Scatter plot: Geographic distribution of top 5 amenity types
# Get top 5 amenities
top_5_amenities = df["amenity"].value_counts().head(5).index.tolist()
df_top = df[df["amenity"].isin(top_5_amenities)].copy()

# Create scatter plot
fig, ax = plt.subplots(figsize=(12, 8))

# Plot each amenity type with different colors/markers
colors = ["red", "blue", "green", "orange", "purple"]
for i, amenity in enumerate(top_5_amenities):
    subset = df_top[df_top["amenity"] == amenity]
    ax.scatter(subset["lon"], subset["lat"], 
               label=amenity, 
               alpha=0.6, 
               s=50,
               color=colors[i])

ax.set_xlabel("Longitude")
ax.set_ylabel("Latitude")
ax.set_title(f"Geographic Distribution of Top 5 Amenities in {CITY_NAME}")
ax.legend(title="Amenity Type", bbox_to_anchor=(1.05, 1), loc="upper left")
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Scatter plot: Data quality vs location
# Create a quality score: count of non-null contact fields (phone, website)
df["contact_quality"] = df[["phone", "website"]].notna().sum(axis=1)

# Create figure with subplots or single plot with color coding
fig, ax = plt.subplots(figsize=(12, 8))

# Scatter by contact quality (0, 1, or 2 fields)
quality_colors = {0: "lightgray", 1: "gold", 2: "darkgreen"}
quality_labels = {0: "No contact info", 1: "Phone or Website", 2: "Phone & Website"}

for quality in [0, 1, 2]:
    subset = df[df["contact_quality"] == quality]
    ax.scatter(subset["lon"], subset["lat"],
               label=quality_labels[quality],
               alpha=0.5,
               s=50,
               color=quality_colors[quality])

ax.set_xlabel("Longitude")
ax.set_ylabel("Latitude")
ax.set_title(f"Data Quality by Location: Contact Info Availability in {CITY_NAME}")
ax.legend(title="Data Quality", loc="best")
ax.grid(True, alpha=0.3)

# Add annotations
total = len(df)
for quality in [0, 1, 2]:
    count = (df["contact_quality"] == quality).sum()
    pct = count / total * 100
    print(f"{quality_labels[quality]}: {count} amenities ({pct:.1f}%)")

plt.tight_layout()
plt.show()
```

```python
# Scatter plot: Language coverage by location
# Analyze name availability in Korean vs English

# Check which name columns exist
has_korean = "name" in df.columns
has_english = "name:en" in df.columns
other_langs = [c for c in df.columns if c.startswith("name:") and c != "name:en"]

print(f"Name columns found: Korean={has_korean}, English={has_english}, Other={other_langs}")

# Create language coverage categories
def categorize_language(row):
    has_name = pd.notna(row.get("name"))
    has_en = pd.notna(row.get("name:en"))
    
    if has_name and has_en:
        return "Korean + English"
    elif has_name and not has_en:
        return "Korean only"
    elif has_en and not has_name:
        return "English only"
    else:
        return "No name"

df["language_coverage"] = df.apply(categorize_language, axis=1)

# Count breakdown
print("\nLanguage coverage breakdown:")
print(df["language_coverage"].value_counts())

# Create scatter plot
fig, ax = plt.subplots(figsize=(12, 8))

# Define colors for each category
lang_colors = {
    "Korean + English": "darkgreen",
    "Korean only": "steelblue",
    "English only": "gold",
    "No name": "lightgray"
}

for category in ["No name", "English only", "Korean only", "Korean + English"]:
    subset = df[df["language_coverage"] == category]
    if len(subset) > 0:
        ax.scatter(subset["lon"], subset["lat"],
                   label=f"{category} ({len(subset)})",
                   alpha=0.6,
                   s=50,
                   color=lang_colors[category])

ax.set_xlabel("Longitude")
ax.set_ylabel("Latitude")
ax.set_title(f"Language Coverage by Location: Name Translations in {CITY_NAME}")
ax.legend(title="Name Availability", loc="best", framealpha=0.9)
ax.grid(True, alpha=0.3)

# Add summary stats
total = len(df)
for category in ["Korean + English", "Korean only", "English only", "No name"]:
    count = (df["language_coverage"] == category).sum()
    pct = count / total * 100
    print(f"{category}: {count} amenities ({pct:.1f}%)")

plt.tight_layout()
plt.show()
```
