```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

sns.set_theme(style="whitegrid")

# TODO: Replace with your city slug and display name
CITY_SLUG = "종로구"       # <-- change this!
CITY_NAME = "종로구"   # <-- for chart titles

df = pd.read_csv(f"../data/clean_{CITY_SLUG}.csv")
print(f"Loaded {len(df)} rows")
```

*My question: Where do different amenity types cluster geographically? Do restaurants/cafés concentrate in specific areas?*

```python
# Scatter plot: Geographic distribution of top 5 amenity types
# Get top 5 amenities
top_5_amenities = df["amenity"].value_counts().head(5).index.tolist()
df_top = df[df["amenity"].isin(top_5_amenities)].copy()

# Create scatter plot
fig, ax = plt.subplots(figsize=(12, 8))

# Plot each amenity type with different colors/markers
colors = ["red", "blue", "green", "orange", "purple"]
for i, amenity in enumerate(top_5_amenities):
    subset = df_top[df_top["amenity"] == amenity]
    ax.scatter(subset["lon"], subset["lat"], 
               label=amenity, 
               alpha=0.6, 
               s=50,
               color=colors[i])

ax.set_xlabel("Longitude")
ax.set_ylabel("Latitude")
ax.set_title(f"Geographic Distribution of Top 5 Amenities in {CITY_NAME}")
ax.legend(title="Amenity Type", bbox_to_anchor=(1.05, 1), loc="upper left")
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Scatter plot: Data quality vs location
# Create a quality score: count of non-null contact fields (phone, website)
df["contact_quality"] = df[["phone", "website"]].notna().sum(axis=1)

# Create figure with subplots or single plot with color coding
fig, ax = plt.subplots(figsize=(12, 8))

# Scatter by contact quality (0, 1, or 2 fields)
quality_colors = {0: "lightgray", 1: "gold", 2: "darkgreen"}
quality_labels = {0: "No contact info", 1: "Phone or Website", 2: "Phone & Website"}

for quality in [0, 1, 2]:
    subset = df[df["contact_quality"] == quality]
    ax.scatter(subset["lon"], subset["lat"],
               label=quality_labels[quality],
               alpha=0.5,
               s=50,
               color=quality_colors[quality])

ax.set_xlabel("Longitude")
ax.set_ylabel("Latitude")
ax.set_title(f"Data Quality by Location: Contact Info Availability in {CITY_NAME}")
ax.legend(title="Data Quality", loc="best")
ax.grid(True, alpha=0.3)

# Add annotations
total = len(df)
for quality in [0, 1, 2]:
    count = (df["contact_quality"] == quality).sum()
    pct = count / total * 100
    print(f"{quality_labels[quality]}: {count} amenities ({pct:.1f}%)")

plt.tight_layout()
plt.show()
```

```python
# Scatter plot: Language coverage by location
# Reload from raw GeoPackage to get name/name:en columns
import geopandas as gpd

gdf_raw = gpd.read_file(f"../data/raw_{CITY_SLUG}.gpkg")
df_raw = pd.DataFrame(gdf_raw.drop(columns=["geometry"]))

# Extract lat/lon
df_raw["lat"] = gdf_raw.geometry.centroid.y
df_raw["lon"] = gdf_raw.geometry.centroid.x

print(f"Loaded {len(df_raw)} raw amenities")
print(f"Name columns available: {[c for c in df_raw.columns if 'name' in c]}")

# Create language coverage categories
def categorize_language(row):
    has_name = pd.notna(row.get("name"))
    has_en = pd.notna(row.get("name:en"))
    
    if has_name and has_en:
        return "Korean + English"
    elif has_name and not has_en:
        return "Korean only"
    elif has_en and not has_name:
        return "English only"
    else:
        return "No name"

df_raw["language_coverage"] = df_raw.apply(categorize_language, axis=1)

# Count breakdown
print("\nLanguage coverage breakdown:")
print(df_raw["language_coverage"].value_counts())

# Create scatter plot
fig, ax = plt.subplots(figsize=(12, 8))

# Define colors for each category
lang_colors = {
    "Korean + English": "darkgreen",
    "Korean only": "steelblue",
    "English only": "gold",
    "No name": "lightgray"
}

for category in ["No name", "English only", "Korean only", "Korean + English"]:
    subset = df_raw[df_raw["language_coverage"] == category]
    if len(subset) > 0:
        ax.scatter(subset["lon"], subset["lat"],
                   label=f"{category} ({len(subset)})",
                   alpha=0.6,
                   s=50,
                   color=lang_colors[category])

ax.set_xlabel("Longitude")
ax.set_ylabel("Latitude")
ax.set_title(f"Language Coverage by Location: Name Translations in {CITY_NAME}")
ax.legend(title="Name Availability", loc="best", framealpha=0.9)
ax.grid(True, alpha=0.3)

# Add summary stats
total = len(df_raw)
print("\nPercentage breakdown:")
for category in ["Korean + English", "Korean only", "English only", "No name"]:
    count = (df_raw["language_coverage"] == category).sum()
    pct = count / total * 100
    print(f"{category}: {count} amenities ({pct:.1f}%)")

plt.tight_layout()
plt.show()
```

```python
# Hexbin plot: Amenity density heatmap
# Shows where amenities cluster geographically

fig, ax = plt.subplots(figsize=(12, 8))

# Create hexbin - shows density of amenities as hexagon colors
hexbin = ax.hexbin(df["lon"], df["lat"], 
                    gridsize=25,  # number of hexagons across
                    cmap="YlOrRd",  # color map: yellow -> orange -> red
                    mincnt=1,  # minimum count to show a hexagon
                    edgecolors="black",
                    linewidths=0.2)

ax.set_xlabel("Longitude")
ax.set_ylabel("Latitude")
ax.set_title(f"Amenity Density Heatmap: {CITY_NAME}")

# Add colorbar showing density
cbar = plt.colorbar(hexbin, ax=ax)
cbar.set_label("Number of Amenities")

plt.tight_layout()
plt.show()

print(f"Hexbin created with {hexbin.get_array().sum():.0f} total amenities visualized")
```

```python
# Create a pastel version of Wistia_r colormap
from matplotlib.colors import ListedColormap, LinearSegmentedColormap
import matplotlib.cm as cm

# Get the original Wistia_r colormap
wistia_cmap = cm.get_cmap("Wistia_r")

# Extract colors and lighten them for pastel effect
colors = [wistia_cmap(x) for x in np.linspace(0, 1, 256)]

# Lighten colors (move toward white) for pastel effect
# Increase lightness by interpolating with white
pastel_colors = []
for rgba in colors:
    r, g, b, a = rgba
    # Mix with white to create pastel (e.g., 60% original + 40% white)
    r_pastel = r * 0.6 + 1.0 * 0.4
    g_pastel = g * 0.6 + 1.0 * 0.4
    b_pastel = b * 0.6 + 1.0 * 0.4
    pastel_colors.append((r_pastel, g_pastel, b_pastel, a))

# Create custom pastel colormap
wistia_pastel = ListedColormap(pastel_colors, name="Wistia_pastel")

print("✓ Pastel Wistia_r colormap created")
```
